```python
import pandas as pd
cut_sentence=pd.read_csv('By_sentence_pos.csv')
cut_sentence['pos']=cut_sentence['pos'].apply(eval)
classiff_ckip_old=pd.read_csv('classiff_ckip.csv')
def word_in_Category(x,cat):
    cat_list=[]
    for word,pos in x:
        word=word.lower().strip()
        if word in classiff_dict[cat]:
            cat_list.append(word)
    if cat_list==[]:
        cat_list=None
    return(cat_list)
classiff_dict={}
for cat in classiff_ckip_old.columns:
    classiff_dict[cat]=list(classiff_ckip_old[classiff_ckip_old[cat].isnull()==False][cat].values)
    cut_sentence[f'word_in_{cat}']=cut_sentence['pos'].apply(lambda x:word_in_Category(x,cat))
```

```python
##情緒分數
opinion_word=pd.read_excel('./emotion_score_Frequecy_filtered.xlsx')
word_dict={}
for w,s in zip(opinion_word['word'],opinion_word['emotion_score']):
    word_dict[w]=s
```

```python
##否定詞
negation=pd.read_csv('./outsourced_data/negation.csv')
negation_list=list(negation.word)
##程度詞,可以利用DfX詞性的詞進行區分 
degree=pd.read_excel('./degree_classif_filtered.xlsx')
degree_dict={}
for strong in degree['strong']:
    degree_dict[strong]=4
for medium in degree['medium']:
    degree_dict[medium]=3
for weak in degree['weak']:
    degree_dict[weak]=2
```

```python
pos_list=['VB','VC','VF','VH','VHC','VI','VL','VJ','VK','A']##情緒分數需要留的詞性
def EMO_Degree(x,pos_list):
    output_list=[]
    for word,pos in x:
        if word in negation_list:
            output_list.append((word,pos,'negation'))
        elif pos in pos_list :
            if word_dict.get(word)!=None:
                output_list.append((word,pos,'emo_score',word_dict.get(word)))
            else:
                output_list.append((word,pos,'other'))
        elif word in degree_dict.keys():
            output_list.append((word,pos,'degree',degree_dict.get(word)))
        else:
            output_list.append((word,pos,'other'))
    return(output_list)
cut_sentence['emo_pos']=cut_sentence['pos'].apply(lambda x:EMO_Degree(x,pos_list))
```

```python
def emotion_score(df):
    total_score=0
    nega_Flag=1
    degree_Flag=1
    count=0
    for emo_pos in df.emo_pos:
        if emo_pos[2]=='emo_score':
            total_score=total_score+(emo_pos[3])*nega_Flag*degree_Flag
        if emo_pos[2]=='degree':
            degree_Flag=emo_pos[3]
        if emo_pos[2]=='negation':
            nega_Flag=-1
        if (emo_pos[2]=='other')&(emo_pos[1]!='DE'):##other or 的
            count+=1
            if count>2: ##間隔不超過兩個字
                nega_Flag=1
                degree_Flag=1
                count=0
    return(total_score)
cut_sentence['emotion_score']=cut_sentence.apply(emotion_score,axis=1)
```

```python
for cat in classiff_ckip_old.columns:
    num=cut_sentence[(cut_sentence[f'word_in_{cat}'].isnull()==False)&(cut_sentence.emotion_score!=0)].shape[0]
    print(cut_sentence[(cut_sentence[f'word_in_{cat}'].isnull()==False)].shape[0])
    print(cat,":",num)
```

```python
def checkPos(x,cate):
    output=[]
    if cate=='emo_score':
        for word in x:
            if word[2]==cate:
                output.append((word[0],word[1],word[3]))

    elif cate=='degree':
        for word in x:
            if word[2]==cate:
                output.append((word[0],word[1],word[3]))
    elif cate=='negation':
        for word in x:
            if word[2]==cate:
                output.append(word[0])
    if output==[]:
        output=None
    return(output)    
cut_sentence['has_emo_Score']=cut_sentence['emo_pos'].apply(lambda x: checkPos(x,"emo_score"))
cut_sentence['has_degree']=cut_sentence['emo_pos'].apply(lambda x: checkPos(x,"degree"))
cut_sentence['has_negation']=cut_sentence['emo_pos'].apply(lambda x: checkPos(x,"negation"))
```

```python
def count_Score(x,flag):
    count=0
    if x!=None:
        if flag>0:
            for w,p,s in x:
                if s>0:
                    count+=1
        else:
            for w,p,s in x:
                if s<0:
                    count+=1
    return(count)
cut_sentence['positive_count']=cut_sentence['has_emo_Score'].apply(lambda x: count_Score(x,1))
cut_sentence['negative_count']=cut_sentence['has_emo_Score'].apply(lambda x: count_Score(x,-1))
```

```python
def count_degree(x,flag):
    count=0
    if x!=None:
        for w,p,s in x:
            if s==flag:
                count+=1
    return(count)
cut_sentence['weak_count']=cut_sentence['has_degree'].apply(lambda x:count_degree(x,2))
cut_sentence['medium_count']=cut_sentence['has_degree'].apply(lambda x:count_degree(x,3))
cut_sentence['strong_count']=cut_sentence['has_degree'].apply(lambda x:count_degree(x,4))
```
