```python
from selenium import webdriver
import time

chrome_path = "C:\selenium_driver_chrome\chromedriver.exe" #chromedriver.exeåŸ·è¡Œæª”æ‰€å­˜åœ¨çš„è·¯å¾‘
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome(chrome_path)

driver.get("https://www.tripadvisor.com.tw/Hotel_Review-g13808853-d455561-Reviews-Cosmos_Hotel_Taipei-Zhongzheng_District_Taipei.html")

```

```python
# è·³é 

nextpage = driver.find_elements_by_xpath("//a[contains(@class,'ui_button nav next primary')]")[0]
totalpage = driver.find_element_by_xpath("//a[contains(text(),'298')]").text


url_list = []
##################################### one time
for i in range(int(totalpage)-1) :
    url_list.append(driver.current_url)
    nextpage.click()
    
    
##################################### one time
```

```python
personal_list=[]
review_list=[]
count = 1

for i in url_list :
    print('page = ',count)
    driver.get(i)

    #è©•è«–è€…è³‡æ–™
    driver.find_element_by_class_name("hotels-review-list-parts-ExpandableReview__ctaLine--1lhYi").click()
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    personal=soup.find_all("div",class_="social-member-event-MemberEventOnObjectBlock__event_wrap--1YkeG")
    #è©•è«–
    reviews=soup.find_all("div",class_="hotels-review-list-parts-SingleReview__mainCol--2XgHm")
    time.sleep(1.5)
    for z in personal:
        try :
            person = z.find_all("div",class_="social-member-MemberHeaderStats__event_info--30wFs")[0]
            place = person.find("span",class_=("social-member-MemberHeaderStats__hometown_stat_item--231iN")).text
        except :
            place = "no_place"
        try :    
            person = z.find_all("div",class_="social-member-MemberHeaderStats__event_info--30wFs")[0]            
            reviewsnum=person.find_all("span",class_=("social-member-MemberHeaderStats__stat_item--34E1r"))[0].text
        except :
            reviewsnum="no_review"
        try :
            person = z.find_all("div",class_="social-member-MemberHeaderStats__event_info--30wFs")[0]            
            allhelpful = person.find_all("span",class_=("social-member-MemberHeaderStats__stat_item--34E1r"))[1].text
        except :
            allhelpful = "no_helpful"
        personal_list.append([place,reviewsnum,allhelpful])    
    for z in reviews:
        title=z.find("div",class_=("hotels-review-list-parts-ReviewTitle__reviewTitle--2Fauz")).text
        review=z.find("div",class_=("cPQsENeY")).text
        rate=z.find("div",class_=("hotels-review-list-parts-RatingLine__bubbles--1oCI4"))
        r=rate.find("span")["class"]
        rating=r[1][-2:]
        try :
            date=z.find("span",class_=("hotels-review-list-parts-EventDate__event_date--CRXs4")).text
        except :
            date = "no_date"
        review_list.append([title,review,rating,date[6:]])
    count += 1
```

```python
print(len(review_list),len(personal_list))
```

```python
import pandas as pd
import numpy as np

title=[]
reviews=[]
rating=[]
place=[]
totalpost=[]
vote=[]

for i in range(len(review_list)):
    title.append(review_list[i][0])
    reviews.append(review_list[i][1])
    rating.append(review_list[i][2])
for i in range(len(personal_list)):
    place.append(personal_list[i][0])
    totalpost.append(personal_list[i][1])
    vote.append(personal_list[i][2])
    

a = pd.DataFrame(title)
b = pd.DataFrame(reviews)
c = pd.DataFrame(rating)
d = pd.DataFrame(place)
e = pd.DataFrame(totalpost)
f = pd.DataFrame(vote)

df =pd.concat([a,b,c,d,e,f],axis=1)
df.columns = ['Title','Reviews','Rating','Place','TotalPost','Vote']
```
