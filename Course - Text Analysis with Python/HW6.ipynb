```python
# beautiful soup
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import requests
import time
from selenium import webdriver

###setting###
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')

#import urllib2
driver = webdriver.Chrome("C:\selenium_driver_chrome\chromedriver.exe", options=options)
driver.get("https://www.booking.com/searchresults.zh-tw.html?aid=397594&label=gog235jc-1DCAEoggI46AdIMFgDaOcBiAEBmAEwuAEXyAEM2AED6AEB-AECiAIBqAIDuAKZtvntBcACAQ&sid=65a1e187908df52e26cdfce936a54714&tmpl=searchresults&checkin_month=11&checkin_monthday=6&checkin_year=2019&checkout_month=11&checkout_monthday=8&checkout_year=2019&class_interval=1&dest_id=-2637882&dest_type=city&dtdisc=0&from_sf=1&group_adults=2&group_children=0&inac=0&index_postcard=0&label_click=undef&no_rooms=1&postcard=0&raw_dest_type=city&room1=A%2CA&sb_price_type=total&shw_aparth=1&slp_r_match=0&src=index&src_elem=sb&srpvid=79e733644df500fd&ss=%E5%8F%B0%E5%8C%97&ss_all=0&ssb=empty&sshis=0&ssne=%E5%8F%B0%E5%8C%97&ssne_untouched=%E5%8F%B0%E5%8C%97&top_ufis=1&rows=25&offset=25")

soup = BeautifulSoup(driver.page_source,"html.parser")
print('ok')
```

```python
#爬取資料，使用for迴圈翻頁
#使用try，except處理缺失值
HotelName = []
HotelPlace = []
HotelRating = []
HotelReviewnum = []
HotelRoom = []
HotelMax = []
HotelPrice = []
for n in range(25):
    url = "https://www.booking.com/searchresults.zh-tw.html?aid=397594&label=gog235jc-1DCAEoggI46AdIMFgDaOcBiAEBmAEwuAEXyAEM2AED6AEB-AECiAIBqAIDuAKZtvntBcACAQ&sid=65a1e187908df52e26cdfce936a54714&tmpl=searchresults&checkin_month=11&checkin_monthday=6&checkin_year=2019&checkout_month=11&checkout_monthday=8&checkout_year=2019&class_interval=1&dest_id=-2637882&dest_type=city&dtdisc=0&from_sf=1&group_adults=2&group_children=0&inac=0&index_postcard=0&label_click=undef&no_rooms=1&postcard=0&raw_dest_type=city&room1=A%2CA&sb_price_type=total&shw_aparth=1&slp_r_match=0&src=index&src_elem=sb&srpvid=79e733644df500fd&ss=%E5%8F%B0%E5%8C%97&ss_all=0&ssb=empty&sshis=0&ssne=%E5%8F%B0%E5%8C%97&ssne_untouched=%E5%8F%B0%E5%8C%97&top_ufis=1&rows=25&offset=" + str(n*25) 
    driver.get(url)
    soup = BeautifulSoup(driver.page_source,"html.parser")
    #飯店名稱
    Name = soup.find_all("span", class_="sr-hotel__name")
    for a in Name:
        k = a.text
        HotelName.append(k[1:-1])
    #飯店地點
    Place = soup.find_all("a", class_="bui-link")
    for a in Place:
        k = a.text
        HotelPlace.append(k[1:5])
    #飯店評分、飯店評語數
    Reviews = soup.find_all("div", class_="reviewFloater reviewFloaterBadge__container")
    for rm in Reviews:
        try:
            num = rm.find_all("div", class_="bui-review-score__text")[0]
            k = num.text
            p = k
        except:
            k = "no review"
            p = k
        try:
            rating = rm.find_all("div", class_="bui-review-score__badge")[0]
            k = rating.text
        except:
            k = "no rating"
        HotelReviewnum.append(p)
        HotelRating.append(k)
    #飯店房間、人數規定、價格
    Rooms = soup.find_all("div", class_="room_details")
    for rm in Rooms:
        try:
            a = rm.find_all("a", class_="room_link")[0]
            k = a.text
            r = k[1:-20]
            m = k[-4:-3]
        except:
            k = "sold out"
            r = k
            m = k
        try:
            b = rm.find_all("div", class_="bui-price-display__value prco-inline-block-maker-helper")[0]
            k = b.text
            p = k[5:-1]
        except:
            k = "sold out"
            p = k
        HotelRoom.append(r)
        HotelMax.append(m)
        HotelPrice.append(p)
```

```python
#檢查筆數是否一致
print(len(HotelName),len(HotelPlace),len(HotelRating),len(HotelReviewnum),len(HotelRoom),len(HotelMax),len(HotelPrice))
```

```python
#整理成DataFrame
import pandas as pd
import numpy as np

a = pd.DataFrame(HotelName)
b = pd.DataFrame(HotelPlace)
c = pd.DataFrame(HotelRating)
d = pd.DataFrame(HotelReviewnum)
e = pd.DataFrame(HotelPrice)
f = pd.DataFrame(HotelRoom)
g = pd.DataFrame(HotelMax)

df = pd.concat([a,b,c,d,e,f,g],axis=1)
df.columns = ['飯店名稱','飯店地區','飯店評分','飯店評論數','飯店價格','飯店房型','飯店房間人數限制']
df
```
