```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import matplotlib
# 直接繪圖，毋須 plot.show()
%matplotlib inline  

# 資料夾設定路徑
path='C:/Users/loy07/Desktop/Python/'
#匯入資料
dfn=pd.read_pickle(path+"df_聯合.pkl")
# 增加年月欄位
ym=[]
for n in range(len(dfn)):
    ym.append(str(dfn.time[n])[:7])
dfn['ym']=ym
#dfn=dfn[dfn['ym']>'2018-02']
dfn
```

```python
##用來確定中美貿易戰開端時間，以第一則新聞日期為指標
dfn1=dfn[(dfn.content.str.contains("中美貿易戰")) ]    # 注意或的符號 |
dfn1.index=range(len(dfn1))
print (len(dfn1))
dfn1
```

```python
#依(中美貿易戰)關鍵字篩出相關文章
USChina_topic=['中美','關稅','台商','貿易戰','製造業','提高','川普','5G','習近平']
find=[]
for s in range(len(dfn)):     
    amount=0 
    for k in USChina_topic:         
        count=dfn.content[s].count(k,0)
        amount+=count
    find.append(amount) 
dfn['中美貿易戰']=find
dfn
```

```python
#新南向政策
Southbound_topic=['新南向','投資','東南亞','東協','移工','區域','人才','華僑','越南','免簽','菲律賓','泰國','馬來西亞','印尼']
find=[]
for s in range(len(dfn)):     
    amount=0 
    for k in Southbound_topic:         
        count=dfn.content[s].count(k,0)
        amount+=count
    find.append(amount) 
dfn['新南向政策']=find
dfn
```

```python
dfn=dfn[dfn['ym']>'2017-12']
month=["2018-01","2018-02","2018-03","2018-04","2018-05","2018-06","2018-07","2018-08","2018-09","2018-10","2018-11","2018-12",
      "2019-01","2019-02","2019-03","2019-04","2019-05","2019-06","2019-07","2019-08","2019-09"]
count=[]
for m in month:
    temp=dfn[(dfn.ym==m)].中美貿易戰.sum()
    count.append(temp)
# 注意要將 type methok 轉成 numeric    
中美貿易戰=pd.to_numeric(count) 

count=[]
for m in month:
    temp=dfn[(dfn.ym==m)].新南向政策.sum()
    count.append(temp)
# 注意要將 type methok 轉成 numeric    
新南向政策=pd.to_numeric(count) 
```

```python
## 時間趨勢
import matplotlib.pyplot as plt 
# 中文設定
from matplotlib import rcParams
rcParams["font.sans-serif"] = ["kaiu"] 
rcParams["font.family"] ="DFKai-sb"

# 加標題，圖例，x軸，y軸
fig, ax = plt.subplots(figsize=(20,10))
# 設定顏色，圖例，線的型態，標記
x=month
#.rolling(5).mean()
ax.plot(x, 中美貿易戰, label='中美貿易戰')       
ax.plot(x, 新南向政策, label='新南向政策')    

# 加年度線
# 年度線的位置
markx=[1,13,25]

# 年度線的高度
marky=1600
for m in markx:
    plt.plot([m,m],[0,1600],'gray') 

# 圖面加文字
text=["2018","2019"]
      
# 加標籤
for i, txt in enumerate(text):
    plt.annotate(txt, [markx[i],marky], size=15)


# 圖例位置由程式自己決定
ax.legend(loc=0);  
ax.set_xlabel('month')     # X label
ax.set_ylabel('count')    # y lobel
# 每隔半年一個標記
ax.set_xticks(np.arange(0,len(month),6))

# 標題
ax.set_title('中美貿易戰與新南向政策 by month');   # title
plt.show()
```

用中美貿易戰與新南向政策看似不相關的兩件事比較，主要是因為想看看會不會因為中美貿易戰有新情況時對台商等有影響，而對台商來說在中國以外可考慮的新據點除了回台便是南洋國家，進而讓我想推測是否政府會對新南向政策做出新的調整計畫。
我認為結果顯示兩者正負相關並不明確，因此我斷定兩者並無太大關聯，而政府面對中美貿易戰台商的做法是以台商回台的方案實行，並無額外為此在推動新南向政策，也可能是新南向政策目的除了協助台商到南向國家投資方展，另一目的是吸引南向國家的人才來台灣，而中美貿易戰下受影響的台商，政府也希望是能夠回台發展促進經濟，兩者目的可都算是希望人才來台，故不會為貿易戰下的台商特別更進南向政策的計畫，因此上述兩者才無關聯性。
