```python
#以句號分割
docsen=doc.split('。')
#刪除\n
sentence=[]
for i in docsen:
    sentence.append(i.replace("\n",""))
del sentence[-1]

#概念定義
#政治號召: ['民主','自由','憲法','法治','人權','民意','選舉','中華民國','政黨','國民黨','民進黨']
#政策:['工資','政策','經濟','市場','貿易','產業','資訊','服務','農','保險','補助','福利']

political_call=['民主','自由','憲法','法治','人權','民意','選舉','中華民國']
policy=['工資','政策','經濟','市場','貿易','產業','資訊','服務','農','保險','補助','福利']
length=[]
keyval1=[]
keyval2=[]
rate1=[]
rate2=[]
for i in sentence:
    length.append(len(i))
    c=0
    d=0
    for a in political_call:
        if i.find(a)>-1:
            c+=1
    for b in policy:
        if i.find(b)>-1:
            d+=1
    keyval1.append(c)
    keyval2.append(d)
    rate1.append(c/len(i))
    rate2.append(d/len(i))

data=list(zip(range(len(sentence)),length,keyval1,keyval2,rate1,rate2))
Tsai=pd.DataFrame(data,columns=['句次','句長','政治號召關鍵字數','政策關鍵字數','政治號召強度','政策強度'])
Tsai
        
```
