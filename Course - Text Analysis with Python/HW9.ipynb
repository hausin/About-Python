```python
path="C:/Python/"
```

```python
import jieba
import re
import pandas as pd
import pickle
```

```python
#讀取excel檔案
hotel_excel=path+'天成飯店評論資料_含日期.xlsx'
df = pd.read_excel(hotel_excel, 0)
df
```

```python
reviews=list(df.Reviews)
reviews
```

```python
#進行資料清理
clear_reviews=[]
for r in reviews:
    r=re.sub("\u3000","",r)
    r=re.sub("\ufeff","",r)
    r=re.sub("，","",r)
    r=re.sub("、","",r)
    r=re.sub("！","",r)
    r=re.sub("。","",r)
    r=re.sub("（","",r)
    r=re.sub("）","",r)
    r=re.sub("\u200d♂️","",r)
    r=re.sub(";","",r)
    r=re.sub("^","",r)
    r=re.sub(":","",r)
    r=re.sub("～","",r)
    r=re.sub("🙂","",r)
    r=re.sub("😄","",r)
    r=re.sub("😌","",r)
    r=re.sub("😊","",r)
    r=re.sub("💁","",r)
    clear_reviews.append(r)

print(clear_reviews)
    
    
```

```python
#建立字典
keyword="'天成','A1','agoda','CP值','M3','翠園','西門町','101','台北車站','國光','南陽街','龍山寺','雙溪','爭鮮','誠品','康是美','瑞芳','7-11','寧夏夜市'"
dict_name=path+'hotel_dict.txt'
with open(dict_name, 'w', encoding='utf8') as f:
    f.write(keyword)
f.close()
```

```python
# 設定斷詞字典
jieba.set_dictionary(path + "dict.txt.big.txt") 
jieba.load_userdict(path + "hotel_dict.txt")
```

```python
#斷詞
token=[]
for i in clear_reviews:
    result = jieba.cut(i)
    text = ' '.join(result)
    token.append(text)
```

```python
#建立新的dataframe
Token=pd.DataFrame(token)
new_df = pd.concat([df.Title,df.Reviews,Token,df.Rating,df.Place,df.Total_Post,df.Vote,df.Date_of_Stay],axis=1)
new_df.columns = ['Title','Reviews','token','Rating','Place','Total_Post','Vote','Date_of_Stay']
```

```python
#存成excel檔
fname = path+'天成飯店評論斷詞.xlsx'
writer = pd.ExcelWriter(fname)
new_df.to_excel(writer,'Sheet1',encoding='utf-8')
writer.save()
```

```python
# 將檔案存成pickle檔
import pickle
with open(path+"Hw9_天成飯店評論斷詞.pkl", "wb") as fp:   #Pickling
     pickle.dump(new_df, fp)

# 讀取檔案(檢查)
checkdata = pd.read_pickle(path+"Hw9_天成飯店評論斷詞.pkl")
checkdata
```
